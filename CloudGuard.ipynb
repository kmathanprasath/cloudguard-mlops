````python
# cloudtrail_parser.py
import json, gzip, glob, pandas as pd

def parse_cloudtrail(folder_path):
    records = []
    for fpath in glob.glob(f'{folder_path}/**/*.json.gz', recursive=True):
        with gzip.open(fpath, 'rt') as f:
            data = json.load(f)
            for r in data.get('Records', []):
                records.append({
                    'timestamp': r.get('eventTime'),
                    'user': r.get('userIdentity', {}).get('userName', 'unknown'),
                    'action': r.get('eventName', ''),
                    'source_ip': r.get('sourceIPAddress', ''),
                    'resource': str(r.get('requestParameters') or ''),
                    'source_type': 'cloudtrail',
                    'label': 0
                })
    return pd.DataFrame(records)

df_ct = parse_cloudtrail('flaws_cloudtrail_logs')
df_ct.to_csv('cloudtrail_parsed.csv', index=False)
print(f"✅ CloudTrail: {len(df_ct)} rows")
print(df_ct.head(3))
````

````python
import pandas as pd

def parse_k8s_beth(filepath):
    df = pd.read_csv(filepath)
    print(f"Columns found: {df.columns.tolist()}")

    records = pd.DataFrame({
        'timestamp': pd.date_range('2024-01-01', periods=len(df), freq='10ms'),
        'user': df.get('userId', 'unknown').astype(str),
        'action': df.get('eventId', '').astype(str),
        'source_ip': '192.168.1.1',  # BETH doesn't have IPs
        'resource': df.get('processId', '').astype(str),
        'source_type': 'k8s',
        'label': df.get('evil', 0)
    })
    return records

df_k8s = parse_k8s_beth('labelled_training_data.csv')
df_k8s.to_csv('k8s_parsed.csv', index=False)
print(f"✅ K8s/BETH: {len(df_k8s)} rows")
print(df_k8s.head(3))
````

````python
import pandas as pd, re

def parse_auth_log(filepath):
    records = []
    pattern = re.compile(
        r'(?P<ts>\w+\s+\d+\s[\d:]+).*?(?P<action>authentication failure|check pass|Accepted|Failed|Invalid)'
        r'.*?rhost=(?P<ip>[\d.]+[\w.-]*)',
        re.IGNORECASE
    )

    with open(filepath, errors='ignore') as f:
        for line in f:
            m = pattern.search(line)
            if m:
                g = m.groupdict()
                action = g['action'].strip()
                records.append({
                    'timestamp': g['ts'],
                    'user': 'unknown',
                    'action': action,
                    'source_ip': g['ip'],
                    'resource': 'ssh',
                    'source_type': 'linux_auth',
                    'label': 0 if 'Accepted' in action else 1
                })

    return pd.DataFrame(records)

df_auth = parse_auth_log('auth_sample.txt')
df_auth.to_csv('linux_auth_parsed.csv', index=False)
print(f"✅ Linux Auth: {len(df_auth)} rows")
print(df_auth['action'].value_counts())
print(df_auth.head(3))
````

````python
import pandas as pd
import numpy as np
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# Step 1 — Load and merge all 3 CSVs
def load_and_fuse():
    df_ct = pd.read_csv('cloudtrail_parsed.csv')
    df_k8s = pd.read_csv('k8s_parsed.csv')
    df_auth = pd.read_csv('linux_auth_parsed.csv')

    combined = pd.concat([df_ct, df_k8s, df_auth], ignore_index=True)
    combined['timestamp'] = pd.to_datetime(combined['timestamp'], errors='coerce')
    combined = combined.dropna(subset=['timestamp'])
    combined = combined.sort_values('timestamp').reset_index(drop=True)

    print(f"✅ Fused dataset: {len(combined)} total rows")
    print(f"   CloudTrail: {len(df_ct)} | K8s: {len(df_k8s)} | Auth: {len(df_auth)}")
    return combined

df = load_and_fuse()
print(df['source_type'].value_counts())
````

````python
def engineer_features(df):
    # Time features
    df['hour'] = df['timestamp'].dt.hour
    df['day_of_week'] = df['timestamp'].dt.dayofweek
    df['is_weekend'] = (df['day_of_week'] >= 5).astype(int)
    df['is_offhours'] = ((df['hour'] < 8) | (df['hour'] > 20)).astype(int)

    # Failed action flag
    df['is_failed'] = df['action'].str.contains(
        'Failed|Denied|Error|Delete|Destroy|failure|invalid',
        case=False, na=False).astype(int)

    # Rare IP flag
    ip_counts = df['source_ip'].value_counts()
    df['is_rare_ip'] = df['source_ip'].map(ip_counts).lt(3).astype(int)

    # Action count per user per hour (simplified)
    df = df.sort_values('timestamp')
    df['action_count_1h'] = df.groupby('user').cumcount() + 1

    # Encode categoricals
    for col in ['user', 'action', 'source_type']:
        le = LabelEncoder()
        df[col + '_enc'] = le.fit_transform(df[col].fillna('unknown').astype(str))

    return df

df = engineer_features(df)
print("✅ Features engineered!")
print(df[['hour','is_weekend','is_offhours','is_failed','is_rare_ip','action_count_1h']].describe())
````

````python
from sklearn.ensemble import IsolationForest
import numpy as np, joblib

X_train = np.load('X_train.npy')
X_test = np.load('X_test.npy')
y_test = np.load('y_test.npy')

print("Training Isolation Forest... (~2 mins)")
model = IsolationForest(
    n_estimators=200,
    contamination=0.005,  # matches our 0.254% attack ratio
    random_state=42,
    n_jobs=-1
)
model.fit(X_train)
joblib.dump(model, 'iforest.pkl')
print("✅ Isolation Forest trained and saved!")
````

````python
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

raw = model.decision_function(X_test)
iforest_scores = 1 - (raw - raw.min()) / (raw.max() - raw.min() + 1e-9)

threshold = 0.6
y_pred = (iforest_scores > threshold).astype(int)

print("=== Isolation Forest Results ===")
print(f"Precision : {precision_score(y_test, y_pred, zero_division=0):.4f}")
print(f"Recall    : {recall_score(y_test, y_pred, zero_division=0):.4f}")
print(f"F1 Score  : {f1_score(y_test, y_pred, zero_division=0):.4f}")
print(f"AUROC     : {roc_auc_score(y_test, iforest_scores):.4f}")
print(f"\nFlagged: {y_pred.sum()} / {len(y_pred)}")
print(f"True attacks in test: {y_test.sum()}")
````

````python
import numpy as np
import tensorflow as tf
from tensorflow import keras

print(f"GPU available: {len(tf.config.list_physical_devices('GPU')) > 0}")

X_train = np.load('X_train.npy')
X_test = np.load('X_test.npy')

SEQ_LEN = 10
N_FEATURES = 10

def make_sequences(X, seq_len=10):
    return np.array([X[i:i+seq_len] for i in range(len(X) - seq_len)])

print("Creating sequences...")
X_seq_train = make_sequences(X_train[:500000])  # use 500K for speed
X_seq_test = make_sequences(X_test[:100000])
print(f"✅ Train sequences: {X_seq_train.shape}")
print(f"✅ Test sequences:  {X_seq_test.shape}")

# Build model
inputs = keras.Input(shape=(SEQ_LEN, N_FEATURES))
x = keras.layers.LSTM(64, return_sequences=True)(inputs)
x = keras.layers.Dropout(0.2)(x)
x = keras.layers.LSTM(16, return_sequences=False)(x)
x = keras.layers.RepeatVector(SEQ_LEN)(x)
x = keras.layers.LSTM(16, return_sequences=True)(x)
x = keras.layers.Dropout(0.2)(x)
x = keras.layers.LSTM(64, return_sequences=True)(x)
outputs = keras.layers.TimeDistributed(keras.layers.Dense(N_FEATURES))(x)

model_lstm = keras.Model(inputs, outputs)
model_lstm.compile(optimizer='adam', loss='mse')
model_lstm.summary()

# Train
history = model_lstm.fit(
    X_seq_train, X_seq_train,
    epochs=30,
    batch_size=256,
    validation_split=0.1,
    callbacks=[keras.callbacks.EarlyStopping(patience=5, restore_best_weights=True)]
)

model_lstm.save('lstm_autoencoder.h5')
print("✅ LSTM Autoencoder trained and saved!")
````

````python
import numpy as np
from sklearn.metrics import precision_score, recall_score, f1_score, roc_auc_score

# Get reconstruction errors on test sequences
print("Getting reconstruction errors...")
X_pred = model_lstm.predict(X_seq_test, batch_size=256)
lstm_scores = np.mean(np.power(X_seq_test - X_pred, 2), axis=(1, 2))

# Align y_test with sequences (sequences start from index 10)
y_test_seq = np.load('y_test.npy')[10:len(X_seq_test)+10]

# Normalize scores to 0-1
lstm_scores_norm = (lstm_scores - lstm_scores.min()) / (lstm_scores.max() - lstm_scores.min() + 1e-9)

print("\n=== LSTM Threshold Finder ===")
print(f"{'Threshold':<12} {'Precision':<12} {'Recall':<10} {'F1':<10} {'Flagged'}")
print("-" * 60)

best_f1 = 0
best_t = 0

for t in [0.3, 0.4, 0.5, 0.6, 0.7, 0.8, 0.9]:
    y_pred = (lstm_scores_norm > t).astype(int)
    p = precision_score(y_test_seq, y_pred, zero_division=0)
    r = recall_score(y_test_seq, y_pred, zero_division=0)
    f = f1_score(y_test_seq, y_pred, zero_division=0)
    print(f"{t:<12} {p:<12.4f} {r:<10.4f} {f:<10.4f} {y_pred.sum()}")
    if f > best_f1:
        best_f1 = f
        best_t = t

print(f"\n✅ Best threshold: {best_t} | Best F1: {best_f1:.4f}")
print(f"AUROC: {roc_auc_score(y_test_seq, lstm_scores_norm):.4f}")
````

````python
import requests, time
from functools import lru_cache

@lru_cache(maxsize=512)
def query_nvd(keyword: str) -> list:
    url = 'https://services.nvd.nist.gov/rest/json/cves/2.0'
    params = {'keywordSearch': keyword, 'resultsPerPage': 5}
    try:
        resp = requests.get(url, params=params, timeout=10)
        resp.raise_for_status()
        data = resp.json()
        cves = []
        for item in data.get('vulnerabilities', []):
            cve = item.get('cve', {})
            metrics = cve.get('metrics', {})
            score = 0.0
            if 'cvssMetricV31' in metrics:
                score = metrics['cvssMetricV31'][0]['cvssData']['baseScore']
            elif 'cvssMetricV2' in metrics:
                score = metrics['cvssMetricV2'][0]['cvssData']['baseScore']
            cves.append({
                'id': cve.get('id'),
                'score': score,
                'desc': cve.get('descriptions', [{}])[0].get('value', '')[:200]
            })
        time.sleep(0.6)
        return cves
    except Exception as e:
        print(f"NVD error: {e}")
        return []

def get_risk_multiplier(cves: list) -> float:
    if not cves: return 1.0
    max_cvss = max(c['score'] for c in cves)
    return 1.0 + (max_cvss / 10.0)

# Test it
print("Testing CVE lookup...")
cves = query_nvd('nginx')
print(f"✅ Found {len(cves)} CVEs for 'nginx'")
for c in cves:
    print(f"   {c['id']} — CVSS: {c['score']} — {c['desc'][:80]}...")
print(f"\nRisk multiplier: {get_risk_multiplier(cves):.2f}")
````

````python
PATTERN_RULES = [
    {
        'name': 'Brute Force',
        'ttp': 'T1110',
        'tactic': 'Credential Access',
        'condition': lambda row: row.get('is_failed') and row.get('action_count_1h', 0) > 20
    },
    {
        'name': 'Cloud Storage Enumeration',
        'ttp': 'T1530',
        'tactic': 'Collection',
        'condition': lambda row: 'List' in str(row.get('action', '')) or
                                 'Describe' in str(row.get('action', ''))
    },
    {
        'name': 'Privilege Escalation',
        'ttp': 'T1078',
        'tactic': 'Privilege Escalation',
        'condition': lambda row: row.get('is_offhours') and row.get('is_rare_ip')
    },
    {
        'name': 'Data Destruction',
        'ttp': 'T1485',
        'tactic': 'Impact',
        'condition': lambda row: 'Delete' in str(row.get('action', '')) or
                                 'Destroy' in str(row.get('action', ''))
    },
    {
        'name': 'Authentication Failure',
        'ttp': 'T1110.001',
        'tactic': 'Credential Access',
        'condition': lambda row: 'failure' in str(row.get('action', '')).lower() or
                                 'Failed' in str(row.get('action', ''))
    },
]

def map_to_attack(row: dict) -> dict:
    matches = [r for r in PATTERN_RULES if r['condition'](row)]
    if not matches:
        return {'ttp': 'T0000', 'tactic': 'Unknown', 'name': 'Unclassified'}
    return matches[0]

# Test it
print("Testing MITRE mapper...")

test1 = {'is_failed': 1, 'action_count_1h': 25, 'action': 'Failed', 'is_offhours': 0, 'is_rare_ip': 0}
test2 = {'action': 'DeleteBucket', 'is_failed': 0, 'action_count_1h': 1, 'is_offhours': 1, 'is_rare_ip': 1}
test3 = {'action': 'ListBuckets', 'is_failed': 0, 'action_count_1h': 5, 'is_offhours': 0, 'is_rare_ip': 0}

for t in [test1, test2, test3]:
    result = map_to_attack(t)
    print(f"✅ Action: {t['action']:<20} → {result['ttp']} {result['name']}")
````

````python
import anthropic, os, json
from google.colab import userdata

# Load Anthropic API key from Colab Secrets
os.environ['ANTHROPIC_API_KEY'] = userdata.get('ANTHROPIC_API_KEY')  # Add ANTHROPIC_API_KEY in Colab Secrets

client = anthropic.Anthropic(api_key=os.environ['ANTHROPIC_API_KEY'])

SYSTEM_PROMPT = '''You are a senior SOC (Security Operations Center) analyst
with 10 years of cloud security experience. You are given a cloud security
anomaly event with its context. Your job is to:
1. Explain in 2-3 plain English sentences WHY this event is suspicious
2. State what attack technique this resembles (if known)
3. Give ONE specific recommended action for the responder

Be concise, direct, and actionable.
Format your response as JSON with keys: explanation, attack_technique, action'''

def explain_anomaly(event: dict, anomaly_score: float, cves: list, ttp: dict) -> dict:
    user_content = f'''
Anomaly Score: {anomaly_score:.3f}

Log event:
{json.dumps(event, indent=2, default=str)}

CVEs found on affected resource:
{json.dumps(cves[:3], indent=2)}

MITRE ATT&CK match:
Technique: {ttp.get("ttp")} - {ttp.get("name")}
Tactic: {ttp.get("tactic")}

Explain this anomaly for a SOC analyst.'''

    response = client.messages.create(
        model='claude-sonnet-4-20250514',
        max_tokens=400,
        system=SYSTEM_PROMPT,
        messages=[{'role': 'user', 'content': user_content}]
    )

    text = response.content[0].text
    try:
        # Clean JSON if wrapped in backticks
        clean = text.strip().replace('```json', '').replace('```', '').strip()
        return json.loads(clean)
    except json.JSONDecodeError:
        return {
            'explanation': text,
            'attack_technique': ttp.get('name'),
            'action': 'Investigate immediately'
        }

print("✅ LLM Explainer ready!")
````
